```python
# create a function to make html to list of strings
#1. make it to a string
#2. remove the craft beers avaible part 
#3. replace <br/> by ','
#4. split into a list 
#5. clean the rows by using split
```

```python
def htmlscraper(html):
    #try:
        #a=str(html).strip('<p><b>Craft beers available:<br/></b>')
    #except:
        #a=str(html).strip('<p><span><b>Craft beers available: </b><br/>')
    b=str(html).replace('<br/>',',')
    alist=b.split(',')
    blist=[]
    for x in alist:
        if  'available' not in x:
            z=x
            if '</span>' in x:
                z = x.strip('</span>')
            elif '<span>' in x:
                z = x.strip('<span>')
            elif "\xa0" in x:
                z = x.split("\xa0")[0]
            elif '</p>' in x:
                z = x.strip('</p>')
            elif '/b><span>' in x:
                z = x.split('/b><span>')[-1]
            blist.append(z)
            
    return blist 
```
