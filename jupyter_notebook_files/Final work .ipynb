```python
def get_brewery(beer):
    driver.get("https://www.beeradvocate.com/search/")
    inputElement = driver.find_element_by_xpath("//form[@name='sF']/input[1]")
    inputElement.clear()
    #inputElement.click()
    inputElement.send_keys(beer)
    #time.sleep(6)
    search= driver.find_element_by_xpath("//form[@name='sF']/input[2]")
    search.click()
    url=driver.current_url
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    if 'beer/profile' in url:
        return soup.find('title').get_text().split('|')[1]
    else:
        for i in soup.find_all('a'):
            if ('beer/profile' in str(i)) and ('span' not in str(i)):
            #print(i)
              
                return i.get_text()
                break
                
    #breweries = soup.find_all('title')
    #brewery=breweries[0].get_text().split('|')[1]
```
